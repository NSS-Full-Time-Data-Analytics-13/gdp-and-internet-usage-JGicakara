Question #9: The median GDP per capita in 2020 was 13,357.99

```python
gdp_decades = gdp_df[(gdp_df["Year"] == 1990) | (gdp_df["Year"] == 2000) | (gdp_df["Year"] == 2010) | (gdp_df["Year"] == 2020)]
gdp_decades
```

```python
plt.figure(figsize = (8,6))
df = gdp_decades
sns.boxplot(x=df["Year"], y=df["GDP_Per_Capita"]);
```

```python
df = gdp_decades
sns.barplot(x=df["Year"], y=df["GDP_Per_Capita"]);
```

```python
sns.regplot(
    x=df["Year"],
    y=df["GDP_Per_Capita"],
    fit_reg=False
)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.regplot(
    x=df["Year"],
    y=df["GDP_Per_Capita"],
    line_kws={"color": "r"},
    ax=ax
)
plt.show()
```
